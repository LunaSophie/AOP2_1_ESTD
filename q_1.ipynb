```python
#EX 1
# Lembrar que o head estará sempre se posicionando como o último item da
# lista e o próximo é o anterior a cabeça, ele se comporta como pilha

class Cliente:
  def __init__(self, nome, valorDaConta):
    self.nome = nome
    self.valorDaConta = valorDaConta

  def __repr__(self):
    return f'{self.nome} -> {self.valorDaConta}'


class Nodo:

  #inicializa um nodo onde o dado é o novo valor do cliente e o proximo é
  #o valor do nodo anterior
  def __init__(self, dado=0, proximoNodo= None):
    self.dado = dado
    self.proximo = proximoNodo

  def __repr__ (self):
    return f' {self.dado} -> {self.proximo}'

  #Não sei porque desse set, pois não está sendo utilizado
  """def setProximo(self, proximoNodo):
      self.proximo= proximoNodo"""


class ListaSimplesmenteEncadeada:
  def __init__(self):
      self.head = None
      self.tamanhoDaLista = 0
      self.soma = 0

  def insere(self, novoDado):
    novoNodo = Nodo(novoDado) #var novoNodo é igual a instancia do Nodo que possui como paramentro novoDado
    novoNodo.proximo = self.head
    self.head = novoNodo
    self.tamanhoDaLista += 1
    self.soma += novoDado.valorDaConta #soma com o valorDaConta que está dentro var novoDado (que vai ser adicionado como cliente)

  def delete(self):
    if self.head == None:
        print('A lista já esta vazia')

    else:
      #pega a soma e subtrai com o valorDaConta do nodo que está no head
      self.soma -= self.head.dado.valorDaConta
      #aponta o head para o nodo anterior, "apagando" o head atual
      self.head = self.head.proximo
      #diminui o tamanho da lista
      self.tamanhoDaLista -= 1
      print('Dado deletado com sucesso')

  def escreveConteudo(self):
    #Escreve o conteúdo da lista somente criado para verificação do formato da lista
    mostrarNodos = ''
    nodoAtual = self.head
    while nodoAtual != None:
      mostrarNodos += f'{nodoAtual.dado}->'
      nodoAtual = nodoAtual.proximo
    return mostrarNodos

  def calcularMedia(self):
    #Pega a soma de toda a lista e divide pelo tamanho da lista e armazena o dado na var media
    media = self.soma/self.tamanhoDaLista
    return media


#Cria lista que irá salvar as instâncias do Nodo
lista = ListaSimplesmenteEncadeada()

while True:
  try:
    print('---------------------------------------')
    comando = str(input('Inserir = i / Deletar = d / Sair = s '))

    if comando == 'i':
      nome = str(input('nome:'))
      valorDaConta = float(input('Valor da compra: R$ '))
      cliente = Cliente(nome, valorDaConta) #instancia um objeto da classe Cliente, guardando os dados do input
      lista.insere(cliente) #chama o método insere usando os dados guardados no obj da classe Cliente

    elif comando == 'd':#chama metodo delete na var lista
      lista.delete()

    elif comando == 's':#saida do loop
      print('Código Finalizado!')
      break

    else:#tratamento de erro caso não seja inserido valor nenhum ou fora das opções
      print('Favor entrar com uma das opções')
      continue

    print('Conteúdo da Lista: ',lista.escreveConteudo())#somente p/ visualização do funcionamento da lista
    print(f'Valor Total: R$ {lista.soma:.2f}')
    print('Registro: ',lista.tamanhoDaLista)
    print(f'Média por Cliente: R$ {lista.calcularMedia():.2f}')

  except:
    continue
```

```python
#EX 2

class Nodo:

  #cria Nodo
  def __init__(self, letra):
    self.letra = letra #letra é cada char da string que entra pelo push.
    self.anterior = None #anterior é para gravar o valor do top anterior

class Pilha:

  #responsável por construir a pilha
  def __init__(self):
    self.top = None
    self._size = 0

  #coloca cada letra da string no nodo
  def push(self, item):
    nodo = Nodo(item)
    nodo.anterior = self.top #aponta o anterior do novo nodo criado para o valor do top
    self.top = nodo #troca o top para o novo nodo criado
    self._size += 1 #aumenta o tamanho da pilha

  def pop(self):
    if self._size > 0: #se existe algo na pilha:
      nodo = self.top #guarda o nodo que será removido
      self.top = self.top.anterior # aponta o top para o nodo anterior, "excluindo" o top atual
      self._size -= 1 #diminui o tamanho da pilha
      return nodo.letra #retorna o char que foi removido


pilha = Pilha()

string = str(input('Colocar string: ')) #entrada da string a ser avaliada
balanceada = True #marcador

for i in string:
  if i == '(': #se o char for igual '(' ele irá dar push e colocar o char na pilha
    pilha.push(i)

  elif i == ')': #caso for igual a ')':
    if pilha.top == None: #se a pilha estiver vazia já sai do loop e afirma que é falsa a afirmação balanceada
      balanceada = False
      break
    else: #se tiver algo na lista irá fazer pop
      pilha.pop()

if pilha.top == None and balanceada == True: #no final para ser considerada balanceada a pilha tem que estar vazia e a afirmação balanceada tem que ser verdadeira
  print('A string é balanceada')
else:
  print('A string não é balanceada')

```
