```python
#EX 1
# Lembrar que a lista está em formato de pilha ou seja, o head estará sempre se
#posicionando como o último item da lista e o próximo é o anterior a cabeça

class Cliente:
    def __init__(self, nome, valorDaConta):
        self.nome = nome
        self.valorDaConta = valorDaConta

    def __repr__(self):
        return f'{self.nome} -> {self.valorDaConta}'


class Nodo:

    def __init__(self, dado=0, proximoNodo= None):
      #criando uma célula para lista onde dado é valor do cliente e o proximo é
      #o valor que entrou
        self.dado = dado
        self.proximo = proximoNodo

    def __repr__ (self):
        return f' {self.dado} -> {self.proximo}'

    """def setProximo(self, proximoNodo):
        self.proximo= proximoNodo"""#Não sei porque desse set, pois esta repetindo o conteúdo do __init__


class ListaSimplesmenteEncadeada:
    def __init__(self):
        self.head = None
        self.tamanhoDaLista = 0
        self.soma = 0

    def insere(self, novoDado):
        novoNodo = Nodo(novoDado) #var novoNodo é a instancia do Nodo que possui como paramentro novoDado
        novoNodo.proximo = self.head
        self.head = novoNodo
        self.tamanhoDaLista += 1
        self.soma += novoDado.valorDaConta #soma com o valorDaConta que está dentro var novoDado (que vai ser adicionado como cliente)

    def delete(self):
        if self.head == None:
            print('A lista já esta vazia')

        else:
            self.soma -= self.head.dado.valorDaConta
            # Pega var soma e subtrai com o var head que possui o valor novoDado anteriormente inserido-> dentro dele pegue o valor
            #dado que está dentro do objeto da classe Nodo -> dentro dele pegue o valor valorDaConta inserido no objeto da classe Cliente
            self.head = self.head.proximo
            #head atual é o head anterior salvo no var proximoNodo ou seja, ele retorna um nodo
            self.tamanhoDaLista -= 1
            print('Dado deletado com sucesso')

    def escreveConteudo(self):
      #Escreve o conteúdo da lista somente criado para verificação do formato da lista
      mostrarNodos = ''
      nodoAtual = self.head
      while nodoAtual != None:
        mostrarNodos += f'{nodoAtual.dado}->'
        nodoAtual = nodoAtual.proximo
      return mostrarNodos

    def calcularMedia(self):
      #Pega a soma de toda a lista e divide pelo tamanho da lista e armazena o dado na var media
      media = self.soma/self.tamanhoDaLista
      return media


lista = ListaSimplesmenteEncadeada()
#Cria var lista que possui as propriedades (head,tamanhoDaLista,soma) e métodos da classe ListaSimplesmenteEncadeada

while True:
  try:
    print('---------------------------------------')
    comando = str(input('Inserir = i / Deletar = d / Sair = s '))

    if comando == 'i':
      nome = str(input('nome:'))
      valorDaConta = float(input('Valor da compra: R$ '))
      cliente = Cliente(nome, valorDaConta) #var cliente guarda os dados do input num obj da classe Cliente
      lista.insere(cliente) #chama o método insere na var lista usando os dados guardado na var cliente
      nodo = Nodo(cliente)#cria var nodo e guarda dados da var cliente num obj da classe Nodo

    elif comando == 'd':
      lista.delete()

    elif comando == 's':
      print('Código Finalizado!')
      break

    else:
      print('Favor entrar com uma das opções')
      continue

    '''print('Conteúdo da Lista: ',lista.escreveConteudo())'''
    print(f'Valor Total: R$ {lista.soma:.2f}')
    print('Registro: ',lista.tamanhoDaLista)
    print(f'Média por Cliente: R$ {lista.calcularMedia():.2f}')

  except:
    continue
```

```python
#EX 2

class Nodo:

    def __init__(self, letra):
      #criando uma célula para pilha onde letra é cada char da string que entra e a
      #anterior é para gravar o dado do top anterior
        self.letra = letra
        self.anterior = None

    def __repr__ (self):
        return f' {self.letra} -> {self.anterior}'


class Pilha:

  def __init__(self):
    #responsável por construir a pilha
    self.top = None
    self._size = 0
    
  def __repr__(self):
    #converte o valor do top para str
    return str(self.top)

  def push(self, item):
    #coloca cada letra da string em nodo
    nodo = Nodo(item) 
    nodo.anterior = self.top #antes de colocar o novo nodo salva o antigo top no anterior 
    self.top = nodo #agora adiciona o novo valor como o top
    self._size += 1 # marca o tamanho da pilha
  
  def pop(self):
    if self._size > 0:
      nodo = self.top #salva no nodo o top
      self.top = self.top.anterior # o valor do top agora é igual ao self.top salvo no anterior
      self._size -= 1
      return nodo.letra
 
      
pilha = Pilha()

string = str(input('Colocar string: '))
count = 0 #somente para marcar

for i in string: #tratamento da string pelo loop
  if i == '(':
    pilha.push(i)

  elif i == ')':
    if pilha.top == None:
      count +=1 #tipo de marcador para avisar que entrou aqui
      break 
    else:
      pilha.pop()
    
if pilha.top == None and count == 0:
  print('A string é balanceada')
else:
  print('A string não é balanceada')

```
